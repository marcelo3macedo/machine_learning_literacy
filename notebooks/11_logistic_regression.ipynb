# Logistic Regression:

- the output variable is categorical and the relationship between the input
variables and the output variable is assumed to be linear.
- used to predict the probability of an event occurring.
- the output is a value between 0 and 1, which can be interpreted as the
probability of the event occurring.
- if the probability is greater than a certain threshold, the event is predicted to
occur; otherwise, it is predicted not to occur.
- commonly used for binary classification problems, where the output variable has
two possible values.

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
# Sample input data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([0, 0, 1, 1, 1])

# Create a logistic regression model
model = LogisticRegression()

# Fit the model to the data
model.fit(X, y)

# Predict the probabilities
X_new = np.array([[6]])
y_prob = model.predict_proba(X_new)

# Print the predicted probabilities
print("Predicted probabilities:", y_prob)
```

```python
# Plot the data points and decision boundary
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Decision Boundary')
plt.scatter(X_new, y_prob[:, 1], color='green', label='Predicted Probability')
plt.xlabel('X')
plt.ylabel('Class')
plt.title('Logistic Regression')
plt.legend()
plt.show()
```
