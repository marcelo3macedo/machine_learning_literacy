# Linear Regression

- the relationship between the input variables and the output variable is assumed
to be linear.
- the goal is to find the best-fit line
- The line is represented by the equation y = mx + b , where y is the output
variable, x is the input variable, m is the slope of the line, and b is the y-
intercept.
- The slope m and the y-intercept b are estimated from the training data using an
optimization algorithm.
- once the line is fitted to the training data, it can be used to make predictions on
new, unseen data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample input data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([3, 4, 7, 11, 13])
```

```python
# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predict the values
X_new = np.array([[6], [7], [8]])
y_pred = model.predict(X_new)

# Print the predicted values
print("Predicted values:", y_pred)
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(X_new, y_pred, color='green', label='Predicted Data')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
