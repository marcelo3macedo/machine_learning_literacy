# Ridge Regression

- adds a penalty term to the regression equation to prevent overfitting of the data.
- the penalty term is a regularization term that penalizes large coefficients
- can help prevent the model from becoming too complex and overfitting the data.
- controlled by a hyperparameter called alpha , which is typically chosen using
cross-validation.
- often used when there are many input variables, and some of them are highly
correlated with each other.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

# Sample input data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

# Create a Ridge regression model with different alpha values
alphas = [0.01, 1, 100]  # Different alpha values
colors = ['blue', 'red', 'green']  # Colors for plotting

# Plot the data points
plt.scatter(X, y, color='black', label='Data Points')

# Fit the models and plot the regression lines
for alpha, color in zip(alphas, colors):
    model = Ridge(alpha=alpha)
    model.fit(X, y)
    y_pred = model.predict(X)
    plt.plot(X, y_pred, color=color, label=f'Ridge (alpha={alpha})')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Ridge Regression with Different Alpha Values')
plt.legend()
plt.show()
```
